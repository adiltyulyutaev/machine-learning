```python
import numpy as np
import re
```

```python
def cos_distance(x,y):

    c = x@y
    x = np.sqrt(x@x)
    y = np.sqrt(y@y)
    c = 1 - c/(x*y)
    return c
```

```python
f = open(r'C:\Users\Пользователь\Desktop\HW1\sentences.txt','r')
```

```python
lines = f.readlines()
lines
```

```python
matrices = []

for line in lines:

    a = [string for string in re.split('[^a-z]',line.lower()) if string != '']
    matrices.append(a)
matrices
```

```python
all_words = list(set([word for sentence in matrices for word in sentence]))
words_index = {word : i for i, word in enumerate(all_words)}
words_index
```

```python
final_matrix = np.zeros((len(lines),len(words_index)))

for i,sentence in enumerate(matrices):
    for word in sentence:
        final_matrix[i,words_index[word]] = sentence.count(word)
final_matrix
```

```python
distances = np.array([cos_distance(final_matrix[0], final_matrix[i]) for i in range(1, final_matrix.shape[0])])
distances
```

```python
near_sentence = np.argsort(distances)[:2] + 1
near_sentence
```
